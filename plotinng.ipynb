## ***Subplot***

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[11,22,33,44,55]
s_hrs=[1,2,3,4,5,6,7,8]
s_score=[11,22,33,22,22,11,6,55]
plt.subplot(1,3,2) #row,colm,index
plt.plot(x,y)

plt.subplot(1,3,1) #row,colm,index
plt.bar(x,y)
plt.subplot(1,3,3) #row,colm,index
plt.scatter(s_hrs,s_score,color='red',marker='<')
plt.show()

```
